```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
diabetes=load_wine()
diabetes
dataset=pd.DataFrame(diabetes.data)
dataset.columns=diabetes.feature_names
dataset.head()
```

```python
#independent and dependent features
x=dataset #input
y=diabetes.target #output feature,dependent
y
```

```python
#train test split
'''This line imports the train_test_split function from the sklearn.model_selection module. This function is used to split arrays or matrices into
random train and test subsets.
x: This is the array or matrix containing the feature data (independent variables).
y: This is the array containing the target data (dependent variable).
test_size=0.30: This parameter specifies the proportion of the dataset to include in the test split. Here, 30% of the data is used for testing,
and the remaining 70% is used for training.random_state=42: This parameter controls the shuffling applied to the data before splitting. 
Providing a random_state ensures reproducibility, so the same split will occur every time the code is run. The number 42 is arbitrary; 
any integer can be used to achieve the same effect.
xtrain: This will contain the training subset of the feature data.
xtest: This will contain the testing subset of the feature data.
ytrain: This will contain the training subset of the target data.
ytest: This will contain the testing subset of the target data.

'''
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=42)
xtrain
```

```python

#standardising the dataset:
'''StandardScaler is a tool from the sklearn.preprocessing module in Python. It helps in making sure all the
features (columns) are on a similar scale.When features in your dataset have very different scales (for example, one feature ranges from 1 to 10 
and another from 1 to 1000), it can cause problems for many machine learning algorithms. These algorithms might give more importance to the feature
with the larger scale, which can lead to poor performance. To avoid this, we standardize the features.
Removing the mean: It subtracts the average value of each feature, so the mean of each feature becomes 0.
Scaling to unit variance: It divides each feature by its standard deviation, so the standard deviation of each feature becomes 1.
Now, both features have a mean of 0 and a standard deviation of 1, making them easier for the machine learning algorithm to process.
Fit: Calculate the mean and standard deviation for each feature in xtrain.
Transform: Use these statistics to scale each feature in xtrain.Here,we are using fit_transform to perform both tasks side by side
'''

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
xtrain=scaler.fit_transform(xtrain)
```

```python
xtest=scaler.transform(xtest)
```

```python
#importing linear regression
from sklearn.linear_model import LinearRegression
#cross validation
from sklearn.model_selection import cross_val_score
regression=LinearRegression()
regression.fit(x,y) #This fits the linear regression model to the training data x (features) and y (target).
mse=cross_val_score(regression,xtrain,ytrain,scoring='neg_mean_squared_error',cv=8)
'''regression: The linear regression model to evaluate.
xtrain: The training data features.
ytrain: The training data target.
scoring='neg_mean_squared_error': This specifies that the evaluation metric is the negative mean squared error (MSE). Cross-validation scores are typically higher is better, so negating the MSE makes it consistent with this convention.
cv=8: This specifies that 8-fold cross-validation should be used.'''
```

```python
np.mean(mse)
```

```python
#prediction
pred=regression.predict(xtest)
```

```python
pred
```

```python
#seaborn library, used for data visualization, top of matplotlib, provides a high-level interface for drawing attractive and informative statistical graphics.
import seaborn as sns
sns.displot(pred-ytest,kind='kde')
# For a good linear regression model, the residuals should be normally distributed and centered around zero. If the KDE plot shows a symmetric, 
#bell-shaped curve centered around zero, it suggests that the model's errors are normally distributed, which is a good sign.
```

```python

'''The r2_score function is used to calculate the coefficient of determination, which is a common metric for evaluating the performance of a
regression model.The R-squared score ranges from 0 to 1, with higher values indicating a better fit of the model to the data.
An R-squared value close to 1 means the model explains most of the variability in the response data.'''
from sklearn.metrics import r2_score
score=r2_score(pred,ytest)
score
```

```python
#LOGISTIC REGRESSION: classification
import pandas as pd
import seaborn as sns
import numpy as np
df=sns.load_dataset('iris')
print(df['species'].unique())
```

```python
#k means clustering
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
x,y=make_blobs(n_samples=1000,centers=3,n_features=2,random_state=23)
x.shape
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
#only interested in x,independnet features,no dependent feature in clustering
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
#manual processs
#elbow method to find value of k
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(xtrain)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
#plot elbow curve
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("no. of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
#value of k is obtained.model will be created
kmeans=KMeans(n_clusters=3,init='k-means++')
ytrain=kmeans.fit_predict(xtrain) #only for training data: both fit nad predict
#for test data only kmeans.predict(xtest)
ytrain
```

```python
xtest
```

```python
#knee locator: find value of k
!pip install kneed
```

```python
from kneed import KneeLocator
kl=KneeLocator(range(1,11),wcss,curve='convex',direction='decreasing')
kl.elbow #convex because value is decreasing of wcss
#no need to follow long method to find value of k,it is shortcut
```

```python
#performance score,silhoutte score
from sklearn.metrics import silhouette_score
sc=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(xtrain)
    score=silhouette_score(xtrain,kmeans.labels_)
    sc.append(score)
    
    
```

```python
sc
```

```python
#check if this sil. score is max for k=3
plt.plot(range(2,11),sc)
plt.xticks(range(2,11))
plt.xlabel("no. of clusters")
plt.ylabel("sc")
plt.show()#max for k=3
```

```python
#hierarichal clustering
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
ir=sns.load_dataset('iris')
ir=ir.drop('species',axis=1)
ir
```

```python
#standardisation
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xs=sc.fit_transform(ir)
xs
```

```python
#applying PCA: convert 4 features to 2
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pcs=pca.fit_transform(xs)
plt.scatter(pcs[:,0],pcs[:,1])
plt.show
```

```python
#agglomerative clustering:making a dandogram
import scipy.cluster.hierarchy as sc
plt.figure(figsize=(20,7))
plt.title("dendograms")
sc.dendrogram(sc.linkage(pcs,method="ward"))
plt.xlabel('sample index')
plt.ylabel("eucledian distance")
```

```python
#selecting longets vertical line such that no horizontal line is passing through it
from sklearn.cluster import AgglomerativeClustering
cl=AgglomerativeClustering(n_clusters=2,metric="euclidean",linkage='ward')
cl.fit(pcs)
cl.labels_
plt.scatter(pcs[:,0],pcs[:,1],c=cl.labels_)
cl.labels_.shape
xtrain.shape
```

```python
from sklearn.metrics import silhouette_score
sc=[]
for k in range(2,11):
    cl=AgglomerativeClustering(n_clusters=k,metric="euclidean",linkage='ward')
    cl.fit(xs)
    score=silhouette_score(xs,cl.labels_)
    sc.append(score)
    
    
```

```python
plt.plot(range(2,11),sc)
plt.xticks(range(2,11))
plt.xlabel("no. of clusters")
plt.ylabel("sc")
plt.show()
```
